### update dataset with KSADS info

###### 1. create a new df using updated df with current dx status
###### 2. add 2 new columns (has dx = has diagnosis, year status event name = baseline data)
###### 3. iterate over rows and check if any value contains a "1" which = has diagnosis 
###### 3.1 and if any row value contains "1", set has_dx to 1
###### 3.2 now set the other cells with 555 or 888 or no value to 0
###### 4. check year status and if baseline arm 1, set year status to 1
###### 4.1 now set the other timepoint row values to 0 

```python
ksads_dx_df=diagnosis_ksads[columns_to_export]
ksads_dx_df["has_dx"]=[None] * len(ksads_dx_df)
ksads_dx_df["year_status"]=[None] * len(ksads_dx_df)

for i, row in ksads_dx_df.iterrows():
    if '1' in row.values:
        ksads_dx_df.at[i,"has_dx"] = 1
    if '1' not in row.values and '555' or '888' in row.values:
        ksads_dx_df.at[i,"has_dx"] = 0
    if 'baseline_year_1_arm_1' in row.values:
        ksads_dx_df.at[i,"year_status"] = 1
    else:
        ksads_dx_df.at[i,"year_status"] = 0


display(ksads_dx_df)
ksads_dx_df.to_csv(os.path.join(export_dir,'updated_with_dx_status_ksads.csv'))
```
